```python
def create_sequences(input_data: pd.DataFrame, target_column, sequence_length):
  sequences = []
  data_size = len(input_data)

  for i in tqdm(range(data_size - sequence_length)):
    sequence = input_data[i:i+sequence_length]

    label_position = i + sequence_length
    label = input_data.iloc[label_position][target_column]

    sequences.append((sequence, label))

  return sequences
```

```python
sequence_length = 1
```

```python
train_sequences = list()
for i in range(len(train_df)):
  train_sequences += (create_sequences(train_df[i], 18, sequence_length=sequence_length))

vali_sequences = list()
for i in range(len(vali_df)):
  vali_sequences += (create_sequences(vali_df[i], 18, sequence_length=sequence_length))
```

# Pytorch Dataset

```python
import random
import os
import math
import argparse
import time
import matplotlib.pyplot as plt

from glob import glob
import cv2
from tqdm.auto import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset

import torchvision.models as models
from torchvision import transforms
```

```python
batch_size = 2
```

```python
class PLANTDataset(Dataset):
  
  def __init__(self, sequences, train_mode=True):
     self.sequences = sequences
     self.train_mode = train_mode

  def __len__(self):
    return len(self.sequences)

  def __getitem__(self, idx):
    if self.train_mode == True:
      sequence, label = self.sequences[idx]
      sequence = torch.tensor(sequence.to_numpy(), dtype=torch.float32)
      label = torch.tensor(label, dtype=torch.float32)
    
      return dict(sequence=sequence, label=label)
    else:
      sequence = self.sequences[idx]
      sequence = torch.tensor(sequence.to_numpy(), dtype=torch.float32)
      return dict(sequence=sequence)
```

```python
train_dataset = PLANTDataset(train_sequences)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle = False, num_workers=0)
```

```python
vali_dataset = PLANTDataset(vali_sequences)
vali_loader = DataLoader(vali_dataset, batch_size=batch_size, shuffle = False, num_workers=0)
```

# Model

```python
n_features = 18
epochs = 30
learning_rate = 0.0001
```

```python
class PricePredictionModel(nn.Module):

  def __init__(self, n_features, n_hidden=128, n_layers=2):
    super().__init__()

    self.n_hidden = n_hidden

    self.lstm = nn.LSTM(
        input_size = n_features,
        hidden_size = n_hidden,
        batch_first = True,
        num_layers = n_layers,
        dropout = 0.2
    )

    self.regressor = nn.Linear(n_hidden, 1)

  def forward(self, x):
    self.lstm.flatten_parameters()

    _, (hidden, _) = self.lstm(x)
    out = hidden[-1]

    return self.regressor(out)
```

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = True
```

```python
seed_everything(27)
```

```python
model = PricePredictionModel(n_features)
model = model.cuda()
optimizer = torch.optim.SGD(params = model.parameters(), lr = learning_rate)
scheduler = None
```

```python
def train(model, optimizer, train_loader, vali_loader, scheduler, device):
  
  criterion = nn.MSELoss()
  best_mae = 9999

  for epoch in range(epochs):
    model.train()
    train_loss = []
    for batch in tqdm(iter(train_loader)):

      sequence, label = batch['sequence'].float().to(device), batch['label'].float().to(device)

      optimizer.zero_grad()

      logit = model(sequence)
      loss = criterion(logit.squeeze(1), label)

      loss.backward()
      optimizer.step()

      train_loss.append(loss.item())

    if scheduler is not None:
      scheduler.step()

    vali_mae = validation(model, vali_loader, criterion, device)

    print(f'Epoch [{epoch}] Train MAE : [{np.mean(train_loss):.5f}] Validation MAE : [{vali_mae:.5f}]\n')
    
    if best_mae > vali_mae:
      best_mae = vali_mae
      torch.save(model.state_dict(), './saved/best_model.pth')
      print('Model Saved')

def validation(model, vali_loader, criterion, device):
  model.eval()
  vali_loss = []
  with torch.no_grad():
    for batch in tqdm(iter(vali_loader)):
      sequence, label = batch['sequence'].float().to(device), batch['label'].float().to(device)

      logit = model(sequence)
      loss = criterion(logit.squeeze(1), label)

      vali_loss.append(loss.item())

  vali_mae_loss = np.mean(vali_loss)
  return vali_mae_loss
```

```python
train(model, optimizer, train_loader, vali_loader, scheduler, device)
```

```python
def evaluate(model, test_loader, device):
    model.eval()
    result = list()
    with torch.no_grad():
        for img in tqdm(iter(test_loader)):
            img = img['sequence'].float().to(device)

            logit = model(img)
            predict = logit.squeeze(1).detach().cpu().tolist()

            result += predict  
                      
    print('test done, saved')  
    return result
       
```

```python
result = evaluate(model, test_loader, device)
```
